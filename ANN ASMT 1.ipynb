Write a Python program to plot a few activation functions that are being used in neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def binaryStep(x):
    return np.heaviside(x,1);

x=np.linspace(-10,10)
plt.plot(x, binaryStep(x))
plt.axis("tight")
plt.title("Binarystep")
plt.show
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x));

x=np.linspace(1,10)
plt.plot(x, sigmoid(x))
plt.axis("tight")
plt.title("Sigmoid")
plt.show;
```

```python
def linear(x):
    return x;

x=np.linspace(1,10)
plt.plot(x, linear(x))
plt.axis("tight")
plt.title("Linear")
plt.show;
```

```python
def tanh(x):
    return np.tanh(x);

x=np.linspace(1,10)
plt.plot(x, tanh(x))
plt.axis("tight")
plt.title("Tanh")
plt.show;
```

```python
def reLU(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1

x=np.linspace(1,10)
plt.plot(x, reLU(x))
plt.axis("tight")
plt.title("RELU")
plt.show;
```

```python
def softMax(x):
    return np.exp(x)/np.sum(np.exp(x), axis=0) 

x=np.linspace(-10,10)
plt.plot(x, softMax(x))
plt.axis("tight")
plt.title("Softmax")
plt.show;
```
